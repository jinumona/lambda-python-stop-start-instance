```python
s = "123abc!@#"

alphaCounter = 0
digitCounter = 0
specialCounter = 0

for char in s:
    
  if char.isdigit():
    
    digitCounter = digitCounter + 1
    
  if char.isalpha():
    
    alphaCounter = alphaCounter + 1
    
  else:
    
    specialCounter = specialCounter + 1
    
    
print(alphaCounter,digitCounter,specialCounter)
```

## stopping zomato dev instance

```python
import boto3
import pprint

ACCESS_KEY = "AKIAXQMPZJZXMP7QBDOE"
SECRET_KEY = "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"
REGION = "ap-south-1"


ec2 = boto3.client('ec2',
                    aws_access_key_id=ACCESS_KEY,
                    aws_secret_access_key=SECRET_KEY,
                    region_name=REGION )

Des_instances = ec2.describe_instances( Filters=[ {'Name':'tag:project', 'Values':["zomato"]},
                                                  {'Name':'tag:env', 'Values':["dev"]}] )


for instance_item in Des_instances['Reservations']:
    
  instance_id = instance_item['Instances'][0]['InstanceId']

  print("Stopping zomato dev Instance Id: {}".format(instance_id))

  ec2.stop_instances(InstanceIds=[ instance_id ])
```

```python
import boto3
import pprint

ACCESS_KEY = "AKIAXQMPZJZXMP7QBDOE"
SECRET_KEY = "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"
REGION = "ap-south-1"


ec2 = boto3.client('ec2',
                    aws_access_key_id=ACCESS_KEY,
                    aws_secret_access_key=SECRET_KEY,
                    region_name=REGION )

#for all dev stopping

Des_instances_dict = ec2.describe_instances( Filters=[ {'Name':'tag:env', 'Values':["dev"]} ] )

# for zomato dev stopping
#Des_instances_dict = ec2.describe_instances( Filters=[ {'Name':'tag:project', 'Values':["zomato"]},
#                                                  {'Name':'tag:env', 'Values':["dev"]}] )


list_Reservations = Des_instances_dict['Reservations']


for reserv_dict in list_Reservations:

  instance_list = reserv_dict['Instances']

  for instance_dict in instance_list:
    
    Inst_id = instance_dict['InstanceId']
    print("Stopping dev instance id:{}".format(Inst_id))
    ec2.stop_instances(InstanceIds=[ Inst_id ])
  
```
